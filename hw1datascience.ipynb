####Scoring/Ranking function to measure general social welfare:

```python
def score(data):    allcosts = list()    l_l = list()    h_l = list()    m_l = list()    w_l = list()            for row in data[1:]:        row = row.replace('unknown','0')        row = row.replace('%', '')        l = row.split(',')[13]        l_l.append(float(l))        h = row.split(',')[14]        h_l.append(float(h))        m = row.split(',')[15]        m_l.append(float(m))        w = row.split(',')[16]        w_l.append(float(w))            l_mean = np.mean(l_l)    h_mean = np.mean(h_l)    m_mean = np.mean(m_l)    w_mean = np.mean(w_l)    l_std = np.std(l_l)    h_std = np.std(h_l)    m_std = np.std(m_l)    w_std = np.std(w_l)            for a,b,c,d in zip(l_l,h_l,m_l,w_l):        cost =  (a-l_mean)/l_std + (b-h_mean)/h_std + (c-m_mean)/m_std + (d-w_mean)/w_std        allcosts.append(cost)            country_list = np.argsort(allcosts)    print    print "Sorted From Worst(At Top) to Best(At Bottom) :"    print    for c in country_list:        print data[c].split(",")[0]        score(data)
```
