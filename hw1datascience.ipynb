####Pairwise correlation, Stonrgest & Weakest Co-relation for each variable , P-value for each Co-relation :

####Linear regression model to predict the average income (GDP per capita) as a function of other variables:

```python
from sklearn import linear_model

def regression(data):
        gdp_list = list()
        all_list = list()             
        for row in data[1:]:
            gdp = row.split(',')[9]
            pop = row.split(',')[7]
            if is_number(gdp) and is_number(pop):
                gdppc = float(gdp)/float(pop)            
                row = row.replace('unknown','0')
                row = row.replace('%', '')
                gdp_list.append(gdppc)
                all_list.append(row.split(',')[10:])          
            
        gdp_list = np.array(gdp_list,dtype='float')
        all_list = np.array(all_list,dtype='float')         
        clf = linear_model.LinearRegression()
        clf.fit(np.array(all_list,dtype='float'),np.array(gdp_list,dtype='float'))
        #print clf.coef_
        
        error_list = list()
        for gdp_element, all_element in zip(gdp_list,all_list):
            s = 0
            for i,j in zip(all_element,clf.coef_):
                s += i*j              
            #print gdp_element, s
            error = (gdp_element-s)
            error_list.append(error)
        
        for country,error in enumerate(error_list):
            if country:
                if error < 0:
                    print "Below Forecast:", data[country].split(",")[0]
                else:
                    print "Above Forecast:", data[country].split(",")[0]

regression(data)
    
```

####Scoring/Ranking function to measure general social welfare:

```python
def score(data):
    allcosts = list()
    l_l = list()
    h_l = list()
    m_l = list()
    w_l = list()
        
    for row in data[1:]:
        row = row.replace('unknown','0')
        row = row.replace('%', '')
        l = row.split(',')[13]  ## Literacy Rate (%)
        l_l.append(float(l))
        h = row.split(',')[14]  ## Health Expenditure/GDP
        h_l.append(float(h))
        m = row.split(',')[15]  ## Military Expenditure/GDP
        m_l.append(float(m))
        w = row.split(',')[16]  ## Renewable Water Resources (cu km)
        w_l.append(float(w))
        
    l_mean = np.mean(l_l)
    h_mean = np.mean(h_l)
    m_mean = np.mean(m_l)
    w_mean = np.mean(w_l)
    l_std = np.std(l_l)
    h_std = np.std(h_l)
    m_std = np.std(m_l)
    w_std = np.std(w_l)
        
    for a,b,c,d in zip(l_l,h_l,m_l,w_l):
        cost =  (a-l_mean)/l_std + (b-h_mean)/h_std + (c-m_mean)/m_std + (d-w_mean)/w_std
        allcosts.append(cost)
        
    country_list = np.argsort(allcosts)
    print
    print "Sorted From Worst(At Top) to Best(At Bottom) :"
    print
    for c in country_list:
        print data[c].split(",")[0]
        
score(data)
```

(a) Scoring Results seems good considerig countries like kenya and namibia appearing top in the list (worst performance) , while United Kingdom and China appearing bottom in the list (very good performance) which is expected behaviour. 
(b) Measurement criteria is very simple and straight forward. It is based on four factors - Literacy Rate (%), Health Expenditure/GDP,   Military Expenditure/GDP and Renewable Water Resources (cu km). Countries performing well in above factors combined have been reported in the results as best in the general social welfare.
